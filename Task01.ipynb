**Visualize the population distribution over the years for a specific country: Aruba**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dataset and skip the 4 rows of metadata
df=pd.read_csv("/content/sample_data/API_SP.POP.TOTL_DS2_en_csv_v2_1040709.csv",skiprows=4)
```

```python
df.head()
```

```python
# Drop the 'Unnamed: 68' column if it exists
if 'Unnamed: 68' in df.columns:
    df.drop(columns=['Unnamed: 68'], inplace=True)
```

```python
# Filter data for a specific country : Aruba
country_data = df[df['Country Name'] == 'Aruba']

# Extract years and population data
years = country_data.columns[4:]  # Assuming the first four columns are metadata
population = country_data.iloc[0, 4:].values
```

```python
# Create a bar chart
plt.figure(figsize=(15, 8))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Aruba Over the Years')
plt.xticks(rotation=90)
plt.show()
```
